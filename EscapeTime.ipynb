```python
import random

 # Función de victoria, pues se han completado los 4 reto del escape room y se llama en el último juego
def escape_completado():
    print("\n\n+----------------------------------------------------------------------+")
    print("|                                                                      |")
    print("|                     ¡EL LABORATORIO QUEDA ATRÁS!                     |")
    print("|                                                                      |")
    print("|         Has superado cada una de las pruebas del científico.         |")
    print("|       Los mecanismos del laboratorio se detienen lentamente...       |")
    print("|            Las puertas de seguridad finalmente se abren.             |")
    print("|                                                                      |")
    print("|       Contra todo pronóstico, lograste escapar del experimento.      |")
    print("|                                                                      |")
    print("|               ¡FELICIDADES! HAS RECUPERADO TU LIBERTAD               |")
    print("|                                                                      |")
    print("+----------------------------------------------------------------------+")

def perdiste(): # Esta función se llama en cada minijuego si has perdido las 3 vidas  y te pregunta si quieres volver a jugar SI/NO
    print("\n\n+----------------------------------------------------------------------+")
    print("|                                                                      |")
    print("|                TE QUEDASTE ATRAPADO EN EL LABORATORIO                |")
    print("|                                                                      |")
    print("|           Las alarmas comienzan a sonar en todo el lugar...          |")
    print("|                    El experimento ha fallado.                        |")
    print("|           Tus errores han agotado todas tus oportunidades.           |")
    print("|                                                                      |")
    print("|     El científico observa en silencio desde la sala de control.      |")
    print("|                                                                      |")
    print("+----------------------------------------------------------------------+")
    print("\nTe has quedado sin vidas; no lograste escapar.\n")
    while True: #Se valida las opciones de SI/NO que el usuario puede ingresar
        jugar = input("¿Volver a intentar desde el inicio? (Si/No): ").strip().lower()
        if jugar in ("si", "sí"):

            print("\n\n+----------------------------------------------------------------------+")
            print("|                        REINICIANDO EXPERIMENTO                       |")
            print("|                                                                      |")
            print("|  Los mecanismos del laboratorio vuelven a activarse lentamente...    |")
            print("|  Cada habitación se prepara para un nuevo intento de escape.         |")
            print("|  ¡Tu ingenio será puesto a prueba nuevamente!                        |")
            print("+----------------------------------------------------------------------+\n\n")
            return True # Este retorno es necesario en el main para seguir jugando
        elif jugar == "no":
            print("\n\n+----------------------------------------------------------------------+")
            print("|                       FIN DEL INTENTO DE ESCAPE                      |")
            print("|                                                                      |")
            print("|  Has decidido abandonar el experimento.                              |")
            print("|  Gracias por participar en esta experiencia de escape.               |")
            print("+----------------------------------------------------------------------+\n")
            return False #Aquí se da por terminada la ejecución del juego
        else:
            print("\n+----------------------------------------------------------------------+")
            print("|                          RESPUESTA NO VÁLIDA                          |")
            print("|                                                                       |")
            print("|         Debes escribir 'Si' para reiniciar o 'No' para salir          |")
            print("+----------------------------------------------------------------------+\n")

def adivinanza(dificultad, vidas): #Primer minijuego de adivinanzas

    ## Versión Fácil

    if dificultad == 1:

        print("\n+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("|                          RETO 1 DESBLOQUEADO                         |")
        print("|                           DIFICULTAD: FÁCIL                          |")
        print("|                                                                      |")
        print("+----------------------------------------------------------------------+")
        print("|                        Prepárate para comenzar...                    |")
        print("|                    Tienes 3 vidas para superar el reto.              |")
        print("|                                                                      |")
        print("|                 Presiona Enter para iniciar el minijuego             |")
        print("+----------------------------------------------------------------------+")
        input()

        print("\n+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("| El primer minijuego es...                                            |")
        print("| Adivinanza, lee con atención el acertijo y escoge la opción correcta:|")
        print("|                                                                      |")
        print("+----------------------------------------------------------------------+")
        print("| ADIVINANZA: LA DESAPARICIÓN DE NOAH                                  |")
        print("+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("| Emmy y Cris tenían un hijo llamado Noah. Un sábado salieron a cenar  |")
        print("| y dejaron al niño en casa.                                           |")
        print("| Al regresar su hijo había desaparecido, preocupados llamaron a la    |")
        print("| policía.                                                             |")
        print("| El detective interrogó a todos en casa:                              |")
        print("|                                                                      |")
        print("| La niñera dijo que había estado preparando la mochila de Noah para   |")
        print("| el día siguiente;                                                    |")
        print("| la empleada dijo que había pasado el día limpiando los baños de      |")
        print("| casa; el cocinero dijo                                               |")
        print("| que había estado en la cocina preparando la cena y no había oído     |")
        print("| nada.                                                                |")
        print("|                                                                      |")
        print("| ¿Quien está mintiendo? (escoja la letra correspondiente a su         |")
        print("| respuesta):                                                          |")
        print("|                                                                      |")
        print("| A. La niñera                                                         |")
        print("| B. La empleada                                                       |")
        print("| C. El cocinero                                                       |")
        print("+----------------------------------------------------------------------+")

        while True: # Valida la entrada del usuario

            resp_adi_facil = input("Respuesta: > ").upper().strip()

            if resp_adi_facil == "A":
                print("\n+----------------------------------------------------------------------+")
                print("|                         ¡Respuesta correcta!                         |")
                print("|                   Has tomado la decisión correcta.                   |")
                print("+----------------------------------------------------------------------+")
                return True, vidas # Este retorna la opción verdadera al main para que continue el siguiente minijuego y se retornan las vidas que quedan

            elif "D" <= resp_adi_facil <= "Z":
                print("\n+----------------------------------------------------------------------+")
                print("|                    No se encuentra en las opciones.                  |")
                print("|                   Solo puedes elegir entre A, B o C.                 |")
                print("+----------------------------------------------------------------------+")

            else:
                vidas -= 1
                print("\n+----------------------------------------------------------------------+")
                print("|                              Incorrecto.                             |")
                print(f"|                         Vidas restantes: {vidas}                           |")
                print("|                  Piensa mejor tu siguiente intento.                  |")
                print("+----------------------------------------------------------------------+")
                break

        if vidas > 0: # Si el usuario todavía tiene vidas y no acertó la anterior adivinanza, se le presenta una diferente

            print("\n+----------------------------------------------------------------------+")
            print("|                                                                      |")
            print("| ADIVINANZA: EL QUE TODO LO VE                                        |")
            print("+----------------------------------------------------------------------+")
            print("|                                                                      |")
            print("| En una casa antigua hay algo que lo ve todo. Está en la sala, en el  |")
            print("| cuarto y hasta en el baño. Nunca habla, pero refleja cada gesto.     |")
            print("|                                                                      |")
            print("| A. Espejo                                                            |")
            print("| B. Cuadro                                                            |")
            print("| C. Ventana                                                           |")
            print("|                                                                      |")
            print("+----------------------------------------------------------------------+")

            while True: # Se repite el proceso de validación con una segunda versión

                resp_adi_facil = input("Respuesta: > ").upper().strip()

                if resp_adi_facil == "A":

                    print("\n+----------------------------------------------------------------------+")
                    print("|                         ¡Respuesta correcta!                         |")
                    print("|                   Has tomado la decisión correcta.                   |")
                    print("+----------------------------------------------------------------------+")
                    return True, vidas

                elif "D" <= resp_adi_facil <= "Z":

                    print("\n+----------------------------------------------------------------------+")
                    print("|                    No se encuentra en las opciones.                  |")
                    print("|                   Solo puedes elegir entre A, B o C.                 |")
                    print("+----------------------------------------------------------------------+")

                else:

                    vidas -= 1

                    print("\n+----------------------------------------------------------------------+")
                    print("|                              Incorrecto.                             |")
                    print(f"|                         Vidas restantes: {vidas}                           |")
                    print("|                  Piensa mejor tu siguiente intento.                  |")
                    print("+----------------------------------------------------------------------+")
                    break

        if vidas > 0: # Si el usuario todavía tiene vidas y no acertó la anterior adivinanza, se le presenta una diferente

            print("\n+----------------------------------------------------------------------+")
            print("|                                                                      |")
            print("| ADIVINANZA: EL BOLSILLO                                              |")
            print("+----------------------------------------------------------------------+")
            print("|                                                                      |")
            print("| Voy en tu bolsillo o bolso, y aunque soy muy pequeñito, te abro      |")
            print("| todas las puertas sin mover un dedito.                               |")
            print("|                                                                      |")
            print("| A. El Inglés                                                         |")
            print("| B. Reloj                                                             |")
            print("| C. La llave                                                          |")
            print("|                                                                      |")
            print("+----------------------------------------------------------------------+")

            while True: # Se repite el proceso de validación con una tercera versión

                resp_adi_facil = input("Respuesta: > ").upper().strip()

                if resp_adi_facil == "C":

                    print("\n+----------------------------------------------------------------------+")
                    print("|                         ¡Respuesta correcta!                         |")
                    print("|                   Has tomado la decisión correcta.                   |")
                    print("+----------------------------------------------------------------------+")
                    return True, vidas

                elif "D" <= resp_adi_facil <= "Z":

                    print("\n+----------------------------------------------------------------------+")
                    print("|                    No se encuentra en las opciones.                  |")
                    print("|                   Solo puedes elegir entre A, B o C.                 |")
                    print("+----------------------------------------------------------------------+")

                else:

                    vidas -= 1

                    print("\n+----------------------------------------------------------------------+")
                    print("|                              Incorrecto.                             |")
                    print(f"|                         Vidas restantes: {vidas}                           |")
                    print("|                  Piensa mejor tu siguiente intento.                  |")
                    print("+----------------------------------------------------------------------+")
                    break
        return False, vidas # Si el usuario pierde las 3 vidas el juego termina

    ## Versión Díficil. Se hace exactamente el mismo proceso, pero con la versión díficil de las adivinanzas
    else:

        print("\n+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("|                          RETO 1 DESBLOQUEADO                         |")
        print("|                          DIFICULTAD: DIFÍCIL                         |")
        print("|                                                                      |")
        print("+----------------------------------------------------------------------+")
        print("|                        Prepárate para comenzar...                    |")
        print("|                    Tienes 3 vidas para superar el reto.              |")
        print("|                                                                      |")
        print("|                 Presiona Enter para iniciar el minijuego             |")
        print("+----------------------------------------------------------------------+")
        input()

        print("\n+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("| ADVININANZA: EL CORREDOR INMÓVIL                                     |")
        print("+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("| ¿Qué es eso que corre todo el día y nunca sale de su sitio?          |")
        print("|                                                                      |")
        print("| A. El reloj                                                          |")
        print("| B. El río                                                            |")
        print("| C. El ventilador                                                     |")
        print("|                                                                      |")
        print("+----------------------------------------------------------------------+")

        while True:

            resp_adi_dificil = input("Respuesta: > ").upper().strip()

            if resp_adi_dificil == "A":

                print("\n+----------------------------------------------------------------------+")
                print("|                         ¡Respuesta correcta!                         |")
                print("|                   Has tomado la decisión correcta.                   |")
                print("+----------------------------------------------------------------------+")
                return True, vidas

            elif "D" <= resp_adi_dificil <= "Z":

                print("\n+----------------------------------------------------------------------+")
                print("|                    No se encuentra en las opciones.                  |")
                print("|                   Solo puedes elegir entre A, B o C.                 |")
                print("+----------------------------------------------------------------------+")

            else:

                vidas -= 1

                print("\n+----------------------------------------------------------------------+")
                print("|                              Incorrecto.                             |")
                print(f"|                         Vidas restantes: {vidas}                           |")
                print("|                  Piensa mejor tu siguiente intento.                  |")
                print("+----------------------------------------------------------------------+")
                break

        if vidas > 0:

            print("\n+----------------------------------------------------------------------+")
            print("|                                                                      |")
            print("| ADIVINANZA: EL ÚLTIMO LECHO                                          |")
            print("+----------------------------------------------------------------------+")
            print("|                                                                      |")
            print("| ¿Qué cosa es que la hacen cantando, la compran llorando y la usan    |")
            print("| sin saber?                                                           |")
            print("|                                                                      |")
            print("| A. Ataúd                                                             |")
            print("| B. El testamento                                                     |")
            print("| C. El féretro funerario                                              |")
            print("|                                                                      |")
            print("+----------------------------------------------------------------------+")

            while True:

                resp_adi_dificil = input("Respuesta: > ").upper().strip()

                if resp_adi_dificil == "A":

                    print("\n+----------------------------------------------------------------------+")
                    print("|                         ¡Respuesta correcta!                         |")
                    print("|                   Has tomado la decisión correcta.                   |")
                    print("+----------------------------------------------------------------------+")
                    return True, vidas

                elif "D" <= resp_adi_dificil <= "Z":

                    print("\n+----------------------------------------------------------------------+")
                    print("|                    No se encuentra en las opciones.                  |")
                    print("|                   Solo puedes elegir entre A, B o C.                 |")
                    print("+----------------------------------------------------------------------+")

                else:

                    vidas -= 1

                    print("\n+----------------------------------------------------------------------+")
                    print("|                              Incorrecto.                             |")
                    print(f"|                         Vidas restantes: {vidas}                           |")
                    print("|                  Piensa mejor tu siguiente intento.                  |")
                    print("+----------------------------------------------------------------------+")
                    break

        if vidas > 0:

            print("\n+----------------------------------------------------------------------+")
            print("|                                                                      |")
            print("| ADIVINANZA: EL DEVORADOR DEL TIEMPO                                  |")
            print("+----------------------------------------------------------------------+")
            print("|                                                                      |")
            print("| Mides mi vida en horas y yo te sirvo expirando. Soy rápido al estar  |")
            print("| delgado y lento al estar gordo.                                      |")
            print("| El viento es mi enemigo. ¿Qué soy?                                   |")
            print("|                                                                      |")
            print("| A. Reloj de arena                                                    |")
            print("| B. El incienso                                                       |")
            print("| C. Vela                                                              |")
            print("|                                                                      |")
            print("+----------------------------------------------------------------------+")

            while True:

                resp_adi_dificil = input("Respuesta: > ").upper().strip()

                if resp_adi_dificil == "C":

                    print("\n+----------------------------------------------------------------------+")
                    print("|                         ¡Respuesta correcta!                         |")
                    print("|                   Has tomado la decisión correcta.                   |")
                    print("+----------------------------------------------------------------------+")
                    return True, vidas

                elif "D" <= resp_adi_dificil <= "Z":

                    print("\n+----------------------------------------------------------------------+")
                    print("|                    No se encuentra en las opciones.                  |")
                    print("|                   Solo puedes elegir entre A, B o C.                 |")
                    print("+----------------------------------------------------------------------+")

                else:

                    vidas -= 1

                    print("\n+----------------------------------------------------------------------+")
                    print("|                              Incorrecto.                             |")
                    print(f"|                         Vidas restantes: {vidas}                           |")
                    print("|                  Piensa mejor tu siguiente intento.                  |")
                    print("+----------------------------------------------------------------------+")
                    break
    return False, vidas

##Inicio Funcion triqui - funciones necesarias

def crear_tablero(): # Se crea la matriz del triqui con listas
    tablero = [
        [" ", " ", " "],
        [" ", " ", " "],
        [" ", " ", " "]
    ]
    return tablero

def mostrar_ayuda(): # Aquí se muestra la asignación de números por casillas para que el usuario entienda

    print("+----------------------------------------------------------------------+")
    print("|                               AYUDA                                  |")
    print("+----------------------------------------------------------------------+")
    print("|                                                                      |")
    print("|  Cada casilla del tablero corresponde a un número del 1 al 9.        |")
    print("|                                                                      |")
    print("|                             1 | 2 | 3                                |")
    print("|                            -----------                               |")
    print("|                             4 | 5 | 6                                |")
    print("|                            -----------                               |")
    print("|                             7 | 8 | 9                                |")
    print("|                                                                      |")
    print("|  Escribe el número donde quieras colocar tu 'X'.                     |")
    print("|                                                                      |")
    print("|  Ejemplo:                                                            |")
    print("|   - '5' marca el centro.                                             |")
    print("|   - '1' marca la esquina superior izquierda.                         |")
    print("|                                                                      |")
    print("|                                                                      |")
    print("|                     Presiona Enter para continuar                    |")
    print("+----------------------------------------------------------------------+")
    input()


posiciones = [ # Se hace un arreglo de las pocisiones del tablero en la matriz triqui
    (0,0), (0,1), (0,2),
    (1,0), (1,1), (1,2),
    (2,0), (2,1), (2,2)
]

def dibujar_tablero(tablero): # Por medio de print se muestra el tablero

    print("+----------------------------------------------------------------------+")
    print("|                              TABLERO                                 |")
    print("+----------------------------------------------------------------------+")
    print("|                                                                      |")
    print(f"|                             {tablero[0][0]} | {tablero[0][1]} | {tablero[0][2]}                                |")
    print("|                            -----------                               |")
    print(f"|                             {tablero[1][0]} | {tablero[1][1]} | {tablero[1][2]}                                |")
    print("|                            -----------                               |")
    print(f"|                             {tablero[2][0]} | {tablero[2][1]} | {tablero[2][2]}                                |")
    print("|                                                                      |")
    print("+----------------------------------------------------------------------+")


def verificar_ganador(tablero, jugador): # Evalua las posibles formas de ganar

    for fila in tablero:
        if fila[0] == fila[1] == fila[2] == jugador: # Evalua si cada fila del tablero esta ocupada por el mismo caracter ( X v O )
            return True

    for col in range(3):
        if tablero[0][col] == tablero[1][col] == tablero[2][col] == jugador: # Evalua si cada columna del tablero esta ocupada por el mismo caracter ( X v O )
            return True

    if tablero[0][0] == tablero[1][1] == tablero[2][2] == jugador: # Evalua si la diagonal derecha del tablero esta ocupada por el mismo caracter ( X v O )
        return True

    if tablero[0][2] == tablero[1][1] == tablero[2][0] == jugador: # Evalua si diagonal izquierda del tablero esta ocupada por el mismo caracter ( X v O )
        return True

    return False # Cuando se retorna a falso es porque el programa no reconoce ninguna combinación de caracteres iguales.


def turno_maquina_facil(tablero): # La maquina usa random para elegir jugadas de forma aleatoria.

    while True:
        numero = random.randint(0, 8)

        fila, col = posiciones[numero]

        if tablero[fila][col] == " ": # Se comprueba que hay un caracter 'vacío' para hacer su jugada.
            tablero[fila][col] = "O"
            break

def turno_maquina_dificil(tablero, posiciones, verificar_ganador):
    # Primero, anotamos qué casillas (del 0 al 8) están vacías.
    # Así no tenemos que buscar en todo el tablero cada vez.
    casillas_vacias = []
    for i in range(9):
        fila, col = posiciones[i]
        if tablero[fila][col] == " ":
            casillas_vacias.append(i)

    # Intentar Ganar
    # La maquina genera una jugada completa 'triqui' de manera estrategica y escoge este camino asumiendo que el usuario no lo va a bloquear.
    for i in casillas_vacias: #La maquina prueba cada casilla vacía.
        fila, col = posiciones[i]
        tablero[fila][col] = "O" #Pongo una O.
        if verificar_ganador(tablero, "O"): #Verifica si hay una jugada ganadora.
            return #Si ganó, se queda con esa jugada y sale de la función.
        tablero[fila][col] = " " #Si no ganó, borra la O y prueba otra.

    # Si la no encuentra una forma de ganar, comienza a bloquear al oponente.
    # Imaginamos qué pasaría si el jugador ('X') juega en esas casillas
    for i in casillas_vacias:
        fila, col = posiciones[i] #La maquina anticipa la jugada del oponente.
        tablero[fila][col] = "X" #Revisa si en esa posición podría ganar el jugador.
        if verificar_ganador(tablero, "X"): #Si el jugador puede ganar lo bloquea.
            tablero[fila][col] = "O" # En esa posición, bloqueamos al jugador
            return #sale de la función
        tablero[fila][col] = " " # Si no gana en esa posición sigue iterando

    #Toma el centro
    #Si la posición del centro (índice 4) está en la lista de vacías, la maquina toma esta posición.
    if 4 in casillas_vacias:
        fila, col = posiciones[4]
        tablero[fila][col] = "O"
        return

    # Si no pudimos ganar, ni bloquear, ni tomar el centro, elegimos una al azar
    if casillas_vacias:
        eleccion_al_azar = random.choice(casillas_vacias)
        fila, col = posiciones[eleccion_al_azar]
        tablero[fila][col] = "O"

##Fin Funcion triqui - funciones necesarias

#Código para ejecutar el juego triqui.

def triqui(dificultad, vidas):

    #Versión fácil

    if dificultad == 1:

        print("\n+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("|                          RETO 2 DESBLOQUEADO                         |")
        print("|                           DIFICULTAD: FÁCIL                          |")
        print("|                                                                      |")
        print("+----------------------------------------------------------------------+")
        print("|                           MINIJUEGO: TRIQUI                          |")
        print("|                                                                      |")
        print("|            Usa los números del 1 al 9 para marcar tu jugada.         |")
        print("|         Marca con X antes de que la máquina complete su línea.       |")
        print("|                                                                      |")
        print("|                     Presiona Enter para comenzar                     |")
        print("+----------------------------------------------------------------------+")
        input()

        while vidas > 0: #Mientras el jugador tenga vidas compite contra la máquina hasta ganar o perder el minijuego.

            tablero = crear_tablero() #llama la función para crear el tablera.

            print("\n+----------------------------------------------------------------------+")
            print("|                       Nuevo tablero iniciado                         |")
            print("|                      Comienza el enfrentamiento                      |")
            print("+----------------------------------------------------------------------+")

            while True:

                dibujar_tablero(tablero) #muestra el tablero del triqui.

                #Turno del jugador
                while True: #evalua la entrada del jugador.
                    try:

                        print("+----------------------------------------------------------------------+")
                        print("|                             TU TURNO                                 |")
                        print("|                 Es momento de marcar tu jugada con X                 |")
                        print("|----------------------------------------------------------------------|")
                        print("|  Elige un número del 1 al 9 según la posición del tablero.           |")
                        print("|  ¿Necesitas orientación? Presiona 0 para ver la guía.                |")
                        print("+----------------------------------------------------------------------+")

                        eleccion = int(input("> "))

                        if eleccion == 0: #Si el jugador no entiende la jugabilidad, se le proporciona una explicación.

                            print("+----------------------------------------------------------------------+")
                            print("|                        Mostrando guía del tablero                    |")
                            print("+----------------------------------------------------------------------+")
                            mostrar_ayuda() #llama a la función con la explicación.
                            dibujar_tablero(tablero)

                        elif 1 <= eleccion <= 9: #Verifica que el número escrito esté entre las opciones.

                            indice = eleccion - 1 # A la elección del jugador se le resta 1 para que coincida con los índices de la matriz del tablero.
                            fila, col = posiciones[indice]

                            if tablero[fila][col] == " ":
                                tablero[fila][col] = "X"
                                print("\n+----------------------------------------------------------------------+")
                                print("|                 Movimiento registrado correctamente                  |")
                                print("+----------------------------------------------------------------------+")
                                break #Si el ingreso en la casilla es válido, sale del while y termina el turno del jugador.
                            else:
                                print("\n+----------------------------------------------------------------------+")
                                print("|              Esa casilla ya está ocupada. Intenta otra.              |")
                                print("+----------------------------------------------------------------------+")

                        else:
                            print("\n+----------------------------------------------------------------------+")
                            print("|             Número fuera de rango. Usa valores del 1 al 9.           |")
                            print("+----------------------------------------------------------------------+")

                    except:
                        print("\n+----------------------------------------------------------------------+")
                        print("|           Entrada inválida. Debes escribir un número válido.         |")
                        print("+----------------------------------------------------------------------+")

                # Verifica si el usuario gana después de su jugada
                if verificar_ganador(tablero, "X"):
                    dibujar_tablero(tablero)
                    print("\n+----------------------------------------------------------------------+")
                    print("|                             ¡GANASTE!                                |")
                    print("|                    Has derrotado a la máquina.                       |")
                    print("|                  Prepárate para el siguiente reto...                 |")
                    print("+----------------------------------------------------------------------+")
                    return True, vidas

                # Verifica empate
                lleno = True #iniciamos suponiendo que hay empate.
                for fila_tablero in tablero: #verificamos si quedan casillas vacias.
                    if " " in fila_tablero:
                        lleno = False #Si quedan casillas vacias entonces no hay empate.

                if lleno: #En caso tal de que lleno siga siendo igual a True, se concluye que hay un empate.
                    dibujar_tablero(tablero)
                    print("\n+----------------------------------------------------------------------+")
                    print("|                               EMPATE                                 |")
                    print("|              Nadie logró dominar el tablero esta ronda.              |")
                    print("|                 El tablero se reinicia... Prepárate.                 |")
                    print("+----------------------------------------------------------------------+")
                    break #Sale de esta ronda e inicia una nueva, no le quita vidas al jugador porque no perdió contra la máquina.


                print("\n+----------------------------------------------------------------------+")
                print("|                        TURNO DE LA MÁQUINA                           |")
                print("|                  La máquina está analizando jugadas...               |")
                print("+----------------------------------------------------------------------+")
                turno_maquina_facil(tablero)

                print("+----------------------------------------------------------------------+")
                print("|                La máquina ha realizado su movimiento.                |")
                print("\n+----------------------------------------------------------------------+")
                # Si la maquina gana, el jugador pierde una vida
                if verificar_ganador(tablero, "O"):
                    vidas -= 1
                    dibujar_tablero(tablero)
                    print("\n+----------------------------------------------------------------------+")
                    print("|                      La máquina ganó esta ronda.                     |")
                    print(f"|                        Vidas restantes: {vidas}                            |")
                    print("|                  No te rindas. Intenta nuevamente.                   |")
                    print("+----------------------------------------------------------------------+")
                    break #Termina esta ronda e inicia una nueva.

                # Se vuelve a verificar si hay empate.
                lleno = True
                for fila_tablero in tablero:
                    if " " in fila_tablero:
                        lleno = False

                if lleno == True:
                    dibujar_tablero(tablero)
                    print("\n+----------------------------------------------------------------------+")
                    print("|                               EMPATE                                 |")
                    print("|                La máquina no logró superarte esta vez.               |")
                    print("|                Se reinicia el tablero para otra ronda.               |")
                    print("+----------------------------------------------------------------------+")
                    break
        return False, vidas # Si el usuario pierde las 3 vidas el juego termina

    #Versión dificil
    else:
        print("\n+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("|                          RETO 2 DESBLOQUEADO                         |")
        print("|                           DIFICULTAD: DIFICIL                        |")
        print("|                                                                      |")
        print("+----------------------------------------------------------------------+")
        print("|                           MINIJUEGO: TRIQUI                          |")
        print("|                                                                      |")
        print("|            Usa los números del 1 al 9 para marcar tu jugada.         |")
        print("|         Marca con X antes de que la máquina complete su línea.       |")
        print("|                                                                      |")
        print("|                     Presiona Enter para comenzar                     |")
        print("+----------------------------------------------------------------------+")
        input()

        while vidas > 0:

            print("+----------------------------------------------------------------------+")
            print("|                       Nuevo tablero iniciado                         |")
            print("|                      Comienza el enfrentamiento                      |")
            print("+----------------------------------------------------------------------+")


            tablero = crear_tablero()

            while True:

                dibujar_tablero(tablero)

                #Turno del jugador.
                while True:
                    try:
                        print("+----------------------------------------------------------------------+")
                        print("|                             TU TURNO                                 |")
                        print("|                 Es momento de marcar tu jugada con X                 |")
                        print("|----------------------------------------------------------------------|")
                        print("|  Elige un número del 1 al 9 según la posición del tablero.           |")
                        print("|  ¿Necesitas orientación? Presiona 0 para ver la guía.                |")
                        print("+----------------------------------------------------------------------+")

                        eleccion = int(input("> "))

                        if eleccion == 0:

                            print("+----------------------------------------------------------------------+")
                            print("|                        Mostrando guía del tablero                    |")
                            print("+----------------------------------------------------------------------+")
                            mostrar_ayuda()
                            dibujar_tablero(tablero)

                        elif 1 <= eleccion <= 9:

                            indice = eleccion - 1
                            fila, col = posiciones[indice]

                            if tablero[fila][col] == " ":
                                tablero[fila][col] = "X"
                                print("\n+----------------------------------------------------------------------+")
                                print("|                 Movimiento registrado correctamente                  |")
                                print("+----------------------------------------------------------------------+")
                                break
                            else:
                                print("\n+----------------------------------------------------------------------+")
                                print("|              Esa casilla ya está ocupada. Intenta otra.              |")
                                print("+----------------------------------------------------------------------+")

                        else:
                            print("\n+----------------------------------------------------------------------+")
                            print("|             Número fuera de rango. Usa valores del 1 al 9.           |")
                            print("+----------------------------------------------------------------------+")

                    except:
                        print("\n+----------------------------------------------------------------------+")
                        print("|           Entrada inválida. Debes escribir un número válido.         |")
                        print("+----------------------------------------------------------------------+")

                if verificar_ganador(tablero, "X"):
                    dibujar_tablero(tablero)
                    print("\n+----------------------------------------------------------------------+")
                    print("|                             ¡GANASTE!                                |")
                    print("|                    Has derrotado a la máquina.                       |")
                    print("|                  Prepárate para el siguiente reto...                 |")
                    print("+----------------------------------------------------------------------+")
                    return True, vidas

                lleno = True
                for fila_tablero in tablero:
                    if " " in fila_tablero:
                        lleno = False

                if lleno:
                    dibujar_tablero(tablero)
                    print("\n+----------------------------------------------------------------------+")
                    print("|                               EMPATE                                 |")
                    print("|              Nadie logró dominar el tablero esta ronda.              |")
                    print("|                 El tablero se reinicia... Prepárate.                 |")
                    print("+----------------------------------------------------------------------+")
                    break

                print("\n+----------------------------------------------------------------------+")
                print("|                        TURNO DE LA MÁQUINA                           |")
                print("|                  La máquina está analizando jugadas...               |")
                print("+----------------------------------------------------------------------+")
                turno_maquina_dificil(tablero, posiciones, verificar_ganador)

                print("+----------------------------------------------------------------------+")
                print("|                La máquina ha realizado su movimiento.                |")
                print("+----------------------------------------------------------------------+")

                if verificar_ganador(tablero, "O"):
                    vidas -= 1
                    dibujar_tablero(tablero)
                    print("\n+----------------------------------------------------------------------+")
                    print("|                      La máquina ganó esta ronda.                     |")
                    print(f"|                        Vidas restantes: {vidas}                            |")
                    print("|                  No te rindas. Intenta nuevamente.                   |")
                    print("+----------------------------------------------------------------------+")
                    break

                lleno = True
                for fila_tablero in tablero:
                    if " " in fila_tablero:
                        lleno = False

                if lleno == True:
                    dibujar_tablero(tablero)
                    print("\n+----------------------------------------------------------------------+")
                    print("|                               EMPATE                                 |")
                    print("|                La máquina no logró superarte esta vez.               |")
                    print("|                Se reinicia el tablero para otra ronda.               |")
                    print("+----------------------------------------------------------------------+")
                    break
        return False, vidas

#Fin Código Principal Triqui

def secuencia(dificultad, vidas): # Se inicia la función del minijuego de secuencia

    ## Versión Fácil

    if dificultad == 1:  # Se escoge la dificultad

        print("\n+----------------------------------------------------------------------+")
        print("|                       TERCER RETO DESBLOQUEADO                       |")
        print("|                        MINIJUEGO 3: SECUENCIA                        |")
        print("|                           DIFICULTAD: FACIL                          |")
        print("+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("|  Ahora el patrón no es tan evidente...                               |")
        print("|  Observa con atención y descubre qué número falta.                   |")
        print("|                                                                      |")
        print("|  Secuencia: 1 - 3 - 6 - 10 - 15 - ?                                  |")
        print("|                                                                      |")
        print("|  Cada error reducirá una vida. Piensa antes de responder.            |")
        print("+----------------------------------------------------------------------+")

        while True: # Se verifica la entrada del usuario

            numeros_ingresados = [] # Se define una lista de numeros ingresados para almacenarlos y que se validen para no repetirse

            for i in range(3): # Se realiza un for para iterar 3 oportunidades en las que el usuario va a ingresar un número

                    try:

                        numero = int(input("\nIngrese el número para completar la secuencia: > "))

                        if numero in numeros_ingresados: # Se revisa si el número ingresado ya fue almacenado
                            print("\n+----------------------------------------------------------------------+")
                            print("|           Ya ingresaste ese número en esta ronda. Intenta otro.      |")
                            print("+----------------------------------------------------------------------+")
                            continue # 'Continue' nos permite preguntar de nuevo el numero a ingresar si el usuario repitio caracter

                        numeros_ingresados.append(numero) # Se apilan los valores ingresados en la lista

                        if numero == 21:
                            print("\n+----------------------------------------------------------------------+")
                            print("|               ¡Correcto! Has descubierto el patrón.                  |")
                            print(f"|                      Vidas restantes: {vidas}                        |")
                            print("|                   Avanzas al siguiente desafío...                    |")
                            print("+----------------------------------------------------------------------+")

                            return True, vidas # Se retorna el 'True' al main y oermite que se continue al siguiente minijuego

                        if numero != 21:
                            vidas -= 1 # Se actualiza el contador de vidas para mostrar al usurio cuantas le quedan
                            print("\n+----------------------------------------------------------------------+")
                            print("|                        Número incorrecto.                            |")
                            print(f"|                    Vidas restantes: {vidas}                           |")
                            print("|                   Avanzas al siguiente desafío...                    |")
                            print("+----------------------------------------------------------------------+")
                            if vidas == 0: # Si no hay más vidas, se retorna al main para acabar el juego
                                return False, vidas

                    except ValueError:
                        print("\n+----------------------------------------------------------------------+")
                        print("|         Entrada inválida. Debes ingresar un número entero.           |")
                        print("+----------------------------------------------------------------------+")

## Versión Dificil
    else:

        print("\n+----------------------------------------------------------------------+")
        print("|                       TERCER RETO DESBLOQUEADO                       |")
        print("|                        MINIJUEGO 3: SECUENCIA                        |")
        print("|                          DIFICULTAD: DIFÍCIL                         |")
        print("+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("|  Ahora el patrón no es tan evidente...                               |")
        print("|  Observa con atención y descubre qué número falta.                   |")
        print("|                                                                      |")
        print("|  Secuencia: 1 - 3 - 4 - 6 - ?                                        |")
        print("|                                                                      |")
        print("|  Cada error reducirá una vida. Piensa antes de responder.            |")
        print("+----------------------------------------------------------------------+")

        while True:

            numeros_ingresados = []


            for i in range(3):

                try:
                    numero= int(input("\nIngrese el número para completar la secuencia: > "))

                    if numero in numeros_ingresados:
                        print("\n+----------------------------------------------------------------------+")
                        print("|     Ya ingresaste ese número en esta ronda. Elige otro diferente.   |")
                        print("+----------------------------------------------------------------------+")
                        continue

                    numeros_ingresados.append(numero)

                    if numero == 4:

                        print("\n+----------------------------------------------------------------------+")
                        print("|               ¡Correcto! Has descubierto el patrón.                  |")
                        print(f"|                    Vidas restantes: {vidas}                           |")
                        print("|                   Avanzas al siguiente desafío...                    |")
                        print("+----------------------------------------------------------------------+")

                        return True, vidas


                    if numero != 4:
                        vidas -= 1
                        print("\n+----------------------------------------------------------------------+")
                        print("|                        Número incorrecto.                            |")
                        print(f"|                    Vidas restantes: {vidas}                                 |")
                        print("|       Analiza mejor la progresión antes de intentar de nuevo.        |")
                        print("+----------------------------------------------------------------------+")
                        if vidas == 0:
                            return False, vidas

                except ValueError:
                    print("\n+----------------------------------------------------------------------+")
                    print("|      Entrada inválida. Debes ingresar un número entero válido.       |")
                    print("+----------------------------------------------------------------------+")

def decision_estrategica(dificultad, vidas): # Se inicia el minijuego final

 #Primero para el nivel fácil
    if dificultad == 1:
        print("\n+----------------------------------------------------------------------+")
        print("|                        CUARTO RETO DESBLOQUEADO                      |")
        print("|                   MINIJUEGO FINAL: DECISIÓN ESTRATÉGICA              |")
        print("|                           DIFICULTAD: FÁCIL                          |")
        print("+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("|  Has llegado al último desafío.                                      |")
        print("|  Aquí no basta con observar… debes elegir con inteligencia.          |")
        print("|                                                                      |")
        print("|  Lee el contexto y selecciona la opción más recomendable.            |")
        print("+----------------------------------------------------------------------+")

        print("\n+----------------------------------------------------------------------+")
        print("|                     RONDA 1: EL ECO DEL ABISMO                       |")
        print("+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("|  Te encuentras ante un abismo con cuatro puentes de cuerdas.         |")
        print("|  Solo uno es totalmente seguro.                                      |")
        print("|                                                                      |")
        print("|  A) Camino A: cruje demasiado, pero es el más corto.                 |")
        print("|  B) Camino B: parece firme, pero la niebla impide ver su final.      |")
        print("|  C) Camino C: piedra sólida, aunque te hace perder tiempo vital.     |")
        print("|  D) Camino D: custodiado por un duende que exige una moneda          |")
        print("|     para dejarte pasar (recurso limitado).                           |")
        print("|                                                                      |")
        print("|  En función del tiempo, integridad y recursos                        |")
        print("|  ¿Qué decisión es la más estratégica?                                |")
        print("+----------------------------------------------------------------------+")

        #Sirve para que si la entrada no es válida vuelva a preguntar por la respuesta.
        while True:

            resp_de_facil = input("Respuesta: > ").upper().strip()

            if resp_de_facil == "D":

                print("\n+----------------------------------------------------------------------+")
                print("|                     ¡Decisión acertada!                              |")
                print("|          Supiste equilibrar riesgo, tiempo y recursos.               |")
                print("|                Has superado el desafío final.                        |")
                print("+----------------------------------------------------------------------+")

                escape_completado()
                return True, vidas

            elif "E" <= resp_de_facil <= "Z":

                print("\n+----------------------------------------------------------------------+")
                print("|            Opción inválida. Solo puedes elegir A, B, C o D.         |")
                print("+----------------------------------------------------------------------+")

            # Si escoje una opción incorrecta, le resta una vida y pasa a la siguiente ronda
            else:

                vidas -= 1

                print("\n+----------------------------------------------------------------------+")
                print("|                       Decisión incorrecta.                           |")
                print(f"|                     Vidas restantes: {vidas}                          |")
                print("|      En situaciones críticas, cada elección tiene consecuencias.     |")
                print("+----------------------------------------------------------------------+")
                break

        #Mientras queden más de 0 vidas repite con las siguientes rondas el proceso
        if vidas > 0:

            print("\n+----------------------------------------------------------------------+")
            print("|                     RONDA 2: LA PRUEBA DEL AGUA                      |")
            print("+----------------------------------------------------------------------+")
            print("|                                                                      |")
            print("|  Viajas solo por la jungla. Una mañana descubres que estás           |")
            print("|  por quedarte sin agua. Solo tienes cuatro opciones para beber.      |")
            print("|                                                                      |")
            print("|  1) Un cactus jugoso del que podrías extraer agua.                   |")
            print("|  2) Un estanque de agua salada.                                      |")
            print("|  3) Un lago transparente y sin movimiento visible.                   |")
            print("|  4) Un arroyo con pequeños sedimentos y minerales.                   |")
            print("|                                                                      |")
            print("|  Consideración importante: tu botella tiene filtro.                  |")
            print("|                                                                      |")
            print("|  ¿Qué opción es la más segura y estratégica?                         |")
            print("+----------------------------------------------------------------------+")

            #Se repite el proceso de validar entrada
            while True:
                try:

                    resp_de_facil = int(input("Respuesta (1-4): > ").strip())

                    if resp_de_facil == 4:

                        print("\n+----------------------------------------------------------------------+")
                        print("|                       ¡Decisión correcta!                            |")
                        print("|         Supiste evaluar el entorno y usar bien tus recursos.         |")
                        print("|                  Has superado la prueba final.                       |")
                        print("+----------------------------------------------------------------------+")

                        escape_completado()
                        return True, vidas

                    elif resp_de_facil < 1 or resp_de_facil > 4:

                        print("\n+----------------------------------------------------------------------+")
                        print("|       El número ingresado no coincide con las opciones (1-4).       |")
                        print("|              Revisa tus alternativas con atención.                  |")
                        print("+----------------------------------------------------------------------+")

                    else:

                        vidas -= 1

                        print("\n+----------------------------------------------------------------------+")
                        print("|                       Elección incorrecta.                          |")
                        print(f"|                     Vidas restantes: {vidas}                          |")
                        print("|   En la supervivencia, una mala decisión puede costar caro.         |")
                        print("+----------------------------------------------------------------------+")

                        break

                except ValueError:

                    print("\n+----------------------------------------------------------------------+")
                    print("|            Entrada inválida. Debes ingresar un número.               |")
                    print("+----------------------------------------------------------------------+")

        #Mientras queden más de 0 vidas repite con las siguientes rondas el proceso
        if vidas > 0:
            print("\n+----------------------------------------------------------------------+")
            print("|               RONDA 3: EL REFUGIO ANTES DE LA TORMENTA               |")
            print("+----------------------------------------------------------------------+")
            print("|                                                                      |")
            print("|  Una tormenta eléctrica se aproxima en 15 minutos.                   |")
            print("|  Estás en campo abierto y debes decidir dónde resguardarte.          |")
            print("|                                                                      |")
            print("|  A) Debajo de un árbol alto y frondoso.                              |")
            print("|  B) Dentro de una cueva poco profunda en una colina.                 |")
            print("|  C) En una estructura metálica abandonada.                           |")
            print("|  D) En un vehículo cerrado con las ventanas arriba.                  |")
            print("|                                                                      |")
            print("|  Considera el riesgo eléctrico y la protección estructural.          |")
            print("|  ¿Cuál es la opción más segura?                                      |")
            print("+----------------------------------------------------------------------+")

            #Se repite el proceso de validar entrada
            while True:

                resp_de_facil = input("Respuesta (A-D): > ").upper().strip()

                if resp_de_facil == "D":

                    print("\n+----------------------------------------------------------------------+")
                    print("|                       ¡Elección correcta!                            |")
                    print("|       Supiste evaluar el riesgo y elegir la opción más segura.       |")
                    print("|                    Has superado la última prueba.                    |")
                    print("+----------------------------------------------------------------------+")

                    escape_completado()
                    return True, vidas

                elif "E" <= resp_de_facil <= "Z":

                    print("\n+----------------------------------------------------------------------+")
                    print("|         Opción inválida. Solo puedes elegir A, B, C o D.            |")
                    print("+----------------------------------------------------------------------+")

                else:

                    vidas -= 1

                    print("\n+----------------------------------------------------------------------+")
                    print("|                        Decisión incorrecta.                          |")
                    print(f"|                      Vidas restantes: {vidas}                         |")
                    print("|           En situaciones críticas, cada elección importa.            |")
                    print("+----------------------------------------------------------------------+")

                    break
            return False, vidas

      #Dificultad difícil
    else:
        print("\n+----------------------------------------------------------------------+")
        print("|                      CUARTO RETO DESBLOQUEADO                        |")
        print("|                 MINIJUEGO FINAL: DECISIÓN ESTRATÉGICA                |")
        print("|                         DIFICULTAD: DIFICIL                          |")
        print("+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("|  Has llegado al último desafío.                                      |")
        print("|  Aquí no basta con observar… debes elegir con inteligencia.          |")
        print("|                                                                      |")
        print("|  Lee el contexto y selecciona la opción más recomendable.            |")
        print("+----------------------------------------------------------------------+")

        print("\n+----------------------------------------------------------------------+")
        print("|                  NIVEL 1: CAMINOS SEPARADOS                          |")
        print("+----------------------------------------------------------------------+")
        print("|                                                                      |")
        print("|  Debes llevar suministros a una ciudad sitiada.                      |")
        print("|  La inteligencia disponible es limitada y cada decisión              |")
        print("|  tendrá impacto directo en la defensa final.                         |")
        print("|                                                                      |")
        print("|  A) Ruta Norte:                                                      |")
        print("|     20% de probabilidad de emboscada, pero el terreno destruye       |")
        print("|     el 50% de la carga.                                              |")
        print("|                                                                      |")
        print("|  B) Ruta Sur:                                                        |")
        print("|     80% de probabilidad de emboscada, pero si logras pasar,          |")
        print("|     la carga llega intacta.                                          |")
        print("|                                                                      |")
        print("|  C) Ruta Central:                                                    |")
        print("|     Sin emboscadas, pero consume el doble de combustible,            |")
        print("|     recurso clave para la defensa final.                             |")
        print("|                                                                      |")
        print("|  ¿Qué ruta es más estratégica considerando probabilidad              |")
        print("|  y logística?                                                        |")
        print("+----------------------------------------------------------------------+")

        while True:

            resp_de_dificil = input("Respuesta (A-C): > ").upper().strip()

            if resp_de_dificil == "A":

                print("\n+----------------------------------------------------------------------+")
                print("|                  ¡Decisión estratégica correcta!                    |")
                print("|       Supiste equilibrar riesgo y pérdida parcial de recursos.      |")
                print("|              La misión puede continuar con éxito.                   |")
                print("+----------------------------------------------------------------------+")

                escape_completado()
                return True, vidas

            elif "E" <= resp_de_dificil <= "Z":

                print("\n+----------------------------------------------------------------------+")
                print("|           Opción inválida. Solo puedes elegir A, B o C.              |")
                print("+----------------------------------------------------------------------+")

            else:

                vidas -= 1

                print("\n+----------------------------------------------------------------------+")
                print("|                        Decisión incorrecta.                         |")
                print(f"|                     Vidas restantes: {vidas}                          |")
                print("|    En estrategia militar, una mala evaluación puede costar caro.    |")
                print("+----------------------------------------------------------------------+")

                break

        if vidas > 0:
            print("\n+----------------------------------------------------------------------+")
            print("|                    NIVEL 2: EL DILEMA DEL REACTOR                    |")
            print("|                       DIFICULTAD: DIFÍCIL                            |")
            print("+----------------------------------------------------------------------+")
            print("|                                                                      |")
            print("|  El reactor energético está al borde del colapso.                    |")
            print("|  Solo puedes ejecutar UNA acción antes de que sea demasiado tarde.   |")
            print("|                                                                      |")
            print("|  A) Reducir potencia al 50%                                          |")
            print("|     Disminuye el riesgo inmediato, pero no elimina la inestabilidad. |")
            print("|                                                                      |")
            print("|  B) Reiniciar completamente el sistema                               |")
            print("|     40% de probabilidad de fallo total.                              |")
            print("|                                                                      |")
            print("|  C) Liberar energía acumulada por válvula externa                    |")
            print("|     Riesgo ambiental moderado, pero estabiliza el núcleo.            |")
            print("|                                                                      |")
            print("|  D) No intervenir y esperar estabilización automática                |")
            print("|     Solo 20% de probabilidad de éxito.                               |")
            print("|                                                                      |")
            print("|  Datos críticos:                                                     |")
            print("|   - Si el reactor falla, la misión termina.                          |")
            print("|   - Si liberas energía, la misión continúa con penalización.         |")
            print("|                                                                      |")
            print("|  ¿Qué decisión maximiza la supervivencia del sistema?                |")
            print("+----------------------------------------------------------------------+")

            while True:

                resp_de_dificil = input("Respuesta (A-D): > ").upper().strip()

                if resp_de_dificil == "C":

                    print("\n+----------------------------------------------------------------------+")
                    print("|                    ¡Decisión táctica correcta!                       |")
                    print("|  Priorizaste la estabilidad del sistema sobre el impacto secundario. |")
                    print("|            El reactor se estabiliza. La misión continúa.             |")
                    print("+----------------------------------------------------------------------+")

                    escape_completado()
                    return True, vidas

                elif "E" <= resp_de_dificil <= "Z":

                    print("\n+----------------------------------------------------------------------+")
                    print("|           Opción inválida. Solo puedes elegir A, B, C o D.           |")
                    print("+----------------------------------------------------------------------+")

                else:

                    vidas -= 1

                    print("\n+----------------------------------------------------------------------+")
                    print("|                     Decisión incorrecta.                             |")
                    print(f"|                   Vidas restantes: {vidas}                            |")
                    print("|        En situaciones críticas, la duda puede ser letal.             |")
                    print("+----------------------------------------------------------------------+")

                break

        if vidas > 0:
            print("\n+----------------------------------------------------------------------+")
            print("|                   NIVEL 3: EVACUACIÓN DEL PUENTE                     |")
            print("|                       DIFICULTAD: DIFÍCIL                            |")
            print("+----------------------------------------------------------------------+")
            print("|                                                                      |")
            print("|  Un puente colgante soporta un máximo de 300 kg.                     |")
            print("|  Debes cruzar con tu equipo sin provocar el colapso.                 |")
            print("|                                                                      |")
            print("|  Peso del equipo:                                                    |")
            print("|   - Tú: 80 kg                                                        |")
            print("|   - Compañero A: 90 kg                                               |")
            print("|   - Compañero B: 70 kg                                               |")
            print("|   - Suministros: 100 kg                                              |")
            print("|                                                                      |")
            print("|  Opciones estratégicas:                                              |")
            print("|   A) Cruzan todos juntos.                                            |")
            print("|   B) Cruzan tres primero y luego uno vuelve por suministros.         |")
            print("|   C) Cruzan dos con suministros y regresan por el resto.             |")
            print("|   D) Transportan los suministros en dos viajes separados.            |")
            print("|                                                                      |")
            print("|  Condición crítica:                                                  |")
            print("|   - Si se supera el peso máximo, el puente colapsa.                  |")
            print("|                                                                      |")
            print("|  ¿Cuál es la decisión más segura y estratégica?                      |")
            print("+----------------------------------------------------------------------+")

            while True:

                resp_de_dificil = input("Respuesta (A-D): > ").upper().strip()

                if resp_de_dificil == "D":

                    print("\n+----------------------------------------------------------------------+")
                    print("|                   ¡Decisión estratégica correcta!                    |")
                    print("|     Calculaste el peso y priorizaste la seguridad del equipo.        |")
                    print("|                  El cruce se realiza sin incidentes.                 |")
                    print("+----------------------------------------------------------------------+")

                    escape_completado()
                    return True, vidas

                elif "E" <= resp_de_dificil <= "Z":

                    print("\n+----------------------------------------------------------------------+")
                    print("|        Opción inválida. Solo puedes elegir A, B, C o D.             |")
                    print("+----------------------------------------------------------------------+")

                else:

                    vidas -= 1

                    print("\n+----------------------------------------------------------------------+")
                    print("|                     Decisión incorrecta.                             |")
                    print(f"|                    Vidas restantes: {vidas}                           |")
                    print("|     Un error de cálculo puede tener consecuencias irreversibles.     |")
                    print("+----------------------------------------------------------------------+")

                    break

        #Sirve en el main para la iteración entre juegos
        return False, vidas

def main():

    print("+----------------------------------------------------------------------+")
    print("|                                                                      |")
    print("|                           ESCAPE TIME                                |")
    print("|                                                                      |")
    print("+----------------------------------------------------------------------+")

    input("\nPresiona Enter para comenzar la historia...")

    print("+----------------------------------------------------------------------+")
    print("|                                                                      |")
    print("|  Un día, tu normalidad se convierte en desesperación.                |")
    print("|  Has sido secuestrado por un científico perturbado por el pasado.    |")
    print("|  Obsesionado con los viajes en el tiempo, te ha convertido en        |")
    print("|  uno de sus experimentos.                                            |")
    print("|                                                                      |")
    print("|  Su plan es enviarte al pasado… y dejarte allí atrapado.             |")
    print("|                                                                      |")
    print("+----------------------------------------------------------------------+")

    input("\nPresiona Enter para continuar...")

    print("+----------------------------------------------------------------------+")
    print("|                                                                      |")
    print("|  Te despiertas dentro de su laboratorio.                             |")
    print("|  Está lleno de pruebas que conducen a la salida.                     |")
    print("|                                                                      |")
    print("|  Solo los más inteligentes logran atravesar las habitaciones         |")
    print("|  y recuperar su libertad.                                            |")
    print("|                                                                      |")
    print("|  ¿Eres capaz de completar los desafíos y escapar?                    |")
    print("|                                                                      |")
    print("+----------------------------------------------------------------------+")

    input("\nPresiona Enter si estás listo para intentarlo...")

    print("+----------------------------------------------------------------------+")
    print("|                                                                      |")
    print("|                  Hoy enfrentarás 4 retos distintos.                  |")
    print("|                                                                      |")
    print("+----------------------------------------------------------------------+")

    print("+----------------------------------------------------------------------+")
    print("|                                                                      |")
    print("|                     Selecciona la dificultad:                        |")
    print("|                                                                      |")
    print("|                             1. Fácil                                 |")
    print("|                             2. Difícil                               |")
    print("|                                                                      |")
    print("+----------------------------------------------------------------------+")

    # Elegir dificultad
    while True:
        try:
            dificultad = int(input("Elige una opción (1 o 2): ").strip())
            if 1 <= dificultad <= 2:
                break
            else:
                print("\nEl número ingresado no coincide con las opciones.")
        except ValueError:
            print("\nOops! No es una entrada válida. Intente de nuevo.")

    jugando = True
    while jugando:
        contador_vidas = 3

        juegos = [adivinanza, triqui, secuencia, decision_estrategica]

        todos_superados = True

        for juego in juegos:

            exito, contador_vidas = juego(dificultad, contador_vidas)

            if exito == False:

                todos_superados = False

                if perdiste() == False:
                    break
                else:
                    break

            elif juego != decision_estrategica:

                print("\n\n+----------------------------------------------------------------------+")
                print("|        ¡Pasaste la prueba! Has escapado de esta habitación...        |")
                print("|             El laboratorio aún guarda más desafíos.                  |")
                print("+----------------------------------------------------------------------+")

        if todos_superados:
            print("\n\n+----------------------------------------------------------------------+")
            print("|                                                                      |")
            print("|                 ¡HAS ESCAPADO DEL LABORATORIO!                        |")
            print("|        El experimento ha fallado… tu mente fue más fuerte.            |")
            print("|                                                                      |")
            print("+----------------------------------------------------------------------+")
            jugando = False
```
